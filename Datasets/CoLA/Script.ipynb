```python
import requests
import pandas as pd
import hub
from hub import transform, schema
import numpy as np
from hub.schema import Primitive
from hub.schema import Text
```

```python
df = pd.read_csv(
    "./cola_public/raw/in_domain_train.tsv",
    sep="\t",
    header=None,
    usecols=[1, 3],
    names=["label", "sentence"],
)
sentences = list(df.sentence.values)
labels = list(df.label.values)
```

```python
data = list(zip(sentences, labels))
```

```python
schema = {
    'sentence': Text(shape=(None, ), max_shape=(500, )),
    'labels': Primitive(dtype="int64")
}
```

```python
@transform(schema=schema)
def load_transform(sample):
    print(sample)
    return {
        "sentence": sample[0],
        "labels": sample[1]
    }
```

```python
ds = load_transform(data)
```
